```python
# Вариант № 4
## Используя рекурсивный метод анализа замкнутых экспоненциальных сетей массового обслуживания, вычислить стационарные характеристики сети N с заданными параметрами.
```

```python
import numpy as np
```

```python
# Решение системы omega*theta=omega с условием нормировки sum(omega)=1
def get_omega(L, theta):
    omega = np.array([1/(L) for _ in range(L)])
    for _ in range(10000):
        n_omega = omega @ theta
        omega = n_omega
    return omega
```

```python
# Условия задачи
L = 8 # число система 
Q = 50 # число требований в закрытой системе
mu = np.array([1, 1.5, 2, 1, 1.3, 1, 1.5, 1.6]) # интенсивности обслуживания 
theta = np.array([ # маршурная матрица
    [0,    0.3,  0,    0.2,  0.2,  0.3,  0,    0],
    [0,    0,    0.4,  0,    0.3,  0,    0.3,  0],
    [0.5,  0,    0,    0.3,  0,    0,    0.1,  0.1],
    [0.2,  0.4,  0,    0,    0,    0.1,  0,    0.3],
    [0,    0,    0.6,  0,    0,    0.4,  0,    0],
    [0,    0,    0,    0.2,  0.2,  0,    0,    0.6],
    [0.3,  0.1,  0,    0,    0,    0.7,  0,    0],
    [0,    0,    0.5,  0,    0,    0,    0.5,  0]
])
```

```python
# Решение уравнений omega*theta = omega с условием нормировки
omega = get_omega(L, theta)
x = omega/mu
```

```python
# Создание массивов для хранения стационарных характеристик
## м.о. числа требований в системах
s = np.zeros((Q+1, L))
## м.о. длительности пребывания требований в системах
u = np.zeros((Q+1, L))
## м.о. длительности реакции сети обслуживания для систем
ksi = np.zeros(L)
## коэффициенты использования систем
psi = np.zeros(L)
## м.о. длительности ожидания требований в очереди системы
w = np.zeros(L)
## м.о. числа требований, ожидающих обслуживание в очереди системы
b = np.zeros(L)
## м.о. числа занятых приборов в системах
h = np.zeros(L)
## интенсивность входящего потока требований в системы
lambdas = np.zeros(L)
## вероятность пребывания требований в системах
p = np.zeros(L)
```

```python
# Вычисление м.о. длительности пребывания требований в системах и м.о. числа требований в системах по формулам (2.1) и (2.2)
for Y in range(1, Q + 1):
    for i in range(L):
        u[Y][i] = (1 / mu[i]) * (s[Y-1][i] + 1)
    for i in range(L):
        summa = 0
        for j in range(L):
            summa += omega[j] * u[Y][j]
        s[Y][i] = omega[i] * u[Y][i] * Y / summa
```

```python
# Вычисление остальных характеристик сети
for i in range(L):
    w[i] = u[Q][i] - (1 / mu[i])
    b[i] = s[Q][i] * w[i] / u[Q][i]
    h[i] = s[Q][i] - b[i]
    lambdas[i] = h[i] * mu[i]
    psi[i] = lambdas[i] / mu[i]
    p[i] = s[Q][i] / Q
    summma = 0
    for j in range(L):
        if i != j:
            summma += omega[j] * u[Q][j]
    ksi[i] = 1 / omega[i] * summa
```

```python
# Вывод результатов
print('\n М.о. длительности реакции сети обслуживания для систем\n')
for i in range(L):
    print(f'Система {i + 1}:\t{ksi[i]}')
print('=' * 120)

print('\nOmega:\n')
for i in range(L):
    print(f'Система {i + 1}:\t{omega[i]}')
print('=' * 120)

print('\nКоэффициенты использования систем:\n')
for i in range(L):
    print(f'Система {i + 1}:\t{psi[i]}')
print('=' * 120)

print('\nМ.о. длительности пребывания требований в системах:\n')
for i in range(L):
    print(f'Система {i + 1}:\t{u[Q][i]}')
print('=' * 120)

print('\nМ.о. длительности ожидания требований в очереди системы:\n')
for i in range(L):
    print(f'Система {i + 1}:\t{w[i]}')
print('=' * 120)

print('\nМ.о. числа требований, ожидающих обслуживание в очереди системы:\n')
for i in range(L):
    print(f'Система {i + 1}:\t{b[i]}')
print('=' * 120)

print('\nМ.о. числа требований в системах:\n')
for i in range(L):
    print(f'Система {i + 1}:\t{s[Q][i]}')
print('=' * 120)

print('\nИнтенсивности потоков требований в системах:\n')
for i in range(L):
    print(f'Система {i + 1}:\t{lambdas[i]}')
print('=' * 120)

print('\nВероятности пребывания требований в системах:\n')
for i in range(L):
    print(f'Система {i + 1}:\t{p[i]}')
print('=' * 120)
```
